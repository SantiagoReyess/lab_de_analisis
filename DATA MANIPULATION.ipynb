```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
data = pd.read_csv("boston.csv").dropna()
```

```python
data
```

```python
data.describe()
```

```python
data_filtered = data.dropna()
```

```python
#Elegimos la columna de y y establecemos n
y = np.reshape(data_filtered['MEDV'], [-1,1])
n = len(y)
```

```python
#Obtenemos todas las columnas del dataframe menos y
data_filtered = data.dropna()
data_filtered = data_filtered.drop(columns = ['AGE', 'INDUS', 'CRIM'])
x_columns = data_filtered.iloc[:,:-1].values
```

```python
#Generamos un vector de unos para StatsModels
ones = np.ones([n, 1])
```

```python
#Concatenamos las columnas de 1 y de x 
X = np.hstack((ones, x_columns))
```

```python
#generamos el modelo
ols = sm.OLS(y,X)
ols_results = ols.fit()
ols_results.summary()
```

```python
columnas = data
columnas = columnas.drop(columns = ['MEDV'])
```

```python
for i in range(len(columnas.columns)):
    minimum = columnas.iloc[:,i].min()
    data_min = columnas.iloc[:,i] - minimum
    maximum = data_min.max()
    data_minimax = data_min / maximum
    columnas.iloc[:, i] = data_minimax
```
