```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuramos el estilo de los gráficos para que tengan una apariencia limpia
sns.set(style="whitegrid")
# Configuramos el tamaño predeterminado de los gráficos
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
df = pd.read_csv("customer_data.csv")
```

#### ANÁLISIS POR SEGMENTO

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
df.groupby('Segment')['NPS_Score'].mean().plot(kind='bar', color='purple')

# Añadimos un título al gráfico
plt.title('Promedio de NPS por Segmento')

# Etiquetamos el eje X
plt.xlabel('Segmento')

# Etiquetamos el eje Y
plt.ylabel('NPS Score')

# Mostramos el gráfico
plt.show()
```

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
df.groupby('Segment')['CAC'].mean().plot(kind='bar', color='skyblue')

# Añadimos un título al gráfico
plt.title('Promedio de CAC por Segmento')

# Etiquetamos el eje X
plt.xlabel('Segmento')

# Etiquetamos el eje Y
plt.ylabel('CAC Promedio')

# Mostramos el gráfico
plt.show()
```

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
df.groupby('Segment')['Expected_LTV_in_Two_Years'].mean().plot(kind='bar', color='yellow')

# Añadimos un título al gráfico
plt.title('Promedio de Expected LTV por Segmento')

# Etiquetamos el eje X
plt.xlabel('Segmento')

# Etiquetamos el eje Y
plt.ylabel('Expected Lifetime Value')

# Mostramos el gráfico
plt.show()
```

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
df.groupby('Segment')['Technical_Issues'].mean().plot(kind='bar', color='orange')

# Añadimos un título al gráfico
plt.title('Promedio de Errores Técnicos por Segmento')

# Etiquetamos el eje X
plt.xlabel('Segmento')

# Etiquetamos el eje Y
plt.ylabel('Promedio de Errores')

# Mostramos el gráfico
plt.show()
```

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
df.groupby('Segment')['Expected_Lifespan_Months'].mean().plot(kind='bar', color='black')

# Añadimos un título al gráfico
plt.title('Promedio Expected Lifespan')

# Etiquetamos el eje X
plt.xlabel('Segmento')

# Etiquetamos el eje Y
plt.ylabel('Expected Lifespan')

# Mostramos el gráfico
plt.show()
```

#### RESPUESTA 1

Si analizamos por segmento, el NPS Score, CAC y el Expected Lifetime Value podemos sacar conclusiones interesantes. Primero, vemos que en promedio, el NPS es más alto para el segmento de Profesionales Jóvenes, seguido por aquellos a media carrera y por útlimo los senior. Además, en promedio, ninguno está por encima del 6, por lo que deberíamos plantear la calidad del servicio y ver qué podemos hacer para elevar el NPS en todos los segmentos. Sin embargo, de ser más valioso un segmento en este aspecto, lo es el de Young Professionals.

Luego, el CAC promedio de los tres segmentos de clientes es muy parecido si no es que prácticamente igual, por lo que en término de costos de adquisición de clientes, lo más eficiente sería comparar con el Expected Lifetime Value, y así poder obtener la relación CAC / Ganancia por cliente. Si analizamos el LTV de cada segmento, en promedio, la ganancia proviene en su mayoría del segmento de Young Professionals.

La gráfica de problemas técnicos nos dice que en promedio, todos los segmentos tienen aproximadamente un error técnico reportado, por lo que podemos descartar esta razón para la poca lealtad a lo largo plazo para todos los segmentos. 

Finalmente, la conclusión más importante, es darnos cuenta que el segmento que más tiempo va a formar parte del negocio es el de Young Professionals, y la relación que tiene con que són los que más ganancias generan respecto a su costo de adquisisción y también son los que más nos promueven.

La estrategia a seguir para mejorar la lealtad de los clientes es analizar cómo ha funcionado la captación de clientes de ese segmento para replicarlo y podamos concentrarnos en seguir captando clientes de este segmento y así mantener la lealtad más altas.

```python
df["LTV/CAC"] = df["Expected_LTV_in_Two_Years"]/df["CAC"]
df
```

```python
# Agrupamos los datos por día de la semana y sumamos las facturas totales por día
# Creamos un gráfico de líneas para visualizar la tendencia de la factura total a lo largo de la semana
ratio = df.groupby(["Acquisition_Month", "Segment"])['LTV/CAC'].mean().reset_index()

# Creamos un gráfico de cajas para visualizar la distribución de la factura total por día de la semana
sns.lineplot(x="Acquisition_Month", y='LTV/CAC', data=ratio, hue="Segment")

# Añadimos un título al gráfico
plt.title('Lifetime Value sobre CAC a través del tiempo')

# Mostramos el gráfico
plt.show()
```
